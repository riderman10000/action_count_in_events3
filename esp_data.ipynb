```python
data = '''0,0,0
1,37,1
2,72,0
3,112,1
4,157,0
5,207,1
6,252,0
7,307,1
8,352,0
9,407,1
10,447,0
11,492,1
12,537,0
13,577,1
14,632,0
15,699,1'''
```

```python
esp_win_change = [int(x.split(',')[-2]) for x in data.split('\n')]
esp_win_change = np.array(esp_win_change)
esp_top_win = [int(x.split(',')[-1]) for x in data.split('\n')]
esp_top_win = np.array(esp_top_win)
```

```python
import matplotlib.pyplot as plt 
plt.plot(df)
current = 0 
start = 0 
for i , (win, trend) in enumerate(zip(esp_win_change, esp_top_win)):
    if i == 0:
        start = win
        current = trend
        continue
    plt.plot(df[start: win], color='red' if current else 'blue')
    start = win 
    current = trend 
```
