```python
data = open('./data.txt', 'r').readlines()
data[2].strip('\n').split(',')
data = data[1:]
data
```

```python
esp_win_change = [int(x.strip('\n').split(',')[-2]) for x in data]
esp_win_change = np.array(esp_win_change)
esp_top_win = [int(x.strip('\n').split(',')[-1]) for x in data]
esp_top_win = np.array(esp_top_win)
```

```python
import matplotlib.pyplot as plt 
plt.plot(df)
current = 0 
start = 0 
for i , (win, trend) in enumerate(zip(esp_win_change, esp_top_win)):
    if i == 0:
        start = win
        current = trend
        continue
    plt.plot(df[start: win], color='red' if current else 'blue')
    start = win 
    current = trend 
```
