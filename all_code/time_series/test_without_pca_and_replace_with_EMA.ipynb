```python
pca_df = dcd.dimensionality_reduction_PCA(manhattan_df)
pca_df.plot()
```

```python
alpha = 0.05
def exponential_moving_average(data, alpha = 0.1):
    ema = [] 
    ema_current = data[0] 
    ema.append(ema_current) 
    
    for value in data[1:] :
        ema_current = alpha * value + (1 - alpha) * ema_current
        ema.append(ema_current) 
    
    return np.array(ema) 

# data = pca_df.values
# smoothed_data = exponential_moving_average(data, alpha) 
```

```python
# import statsmodels.api as sm 
# _, smooth_x = sm.tsa.filters.hpfilter(manhattan_df['x']) 
# _, smooth_y = sm.tsa.filters.hpfilter(manhattan_df['y']) 


smooth_x = exponential_moving_average(manhattan_df['x'], alpha)
smooth_y = exponential_moving_average(manhattan_df['y'], alpha)

smooth_x = smooth_x-smooth_x.mean()
smooth_y = smooth_y -smooth_y.mean()

pd.DataFrame(smooth_x).plot()
pd.DataFrame(smooth_y).plot()
```
