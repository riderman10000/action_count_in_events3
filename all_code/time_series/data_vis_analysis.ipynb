```python
import os 
import glob 
import pandas as pd 
import matplotlib.pyplot as plt 
```

# analysis of natural data

```python
natural_data_path = "../../event_csv/split_data/class2/user02_lab.csv"
data_df = pd.read_csv(natural_data_path, dtype=int)
data_df.plot(x='timestamp', y=['x', 'y'])
data_df.describe()
```

```python

data_df = None 
all_data_df = None 

natural_data_path = "../../event_csv/split_data/class2/*.csv"

# creating subplots 
data_rows = len(glob.glob(natural_data_path))
fig, ax = plt.subplots(data_rows, 1)

for (ax, path) in zip(ax.flatten(), glob.glob(natural_data_path)):
    print(path)
    data_df = pd.read_csv(path)
    data_df.plot( y=['x', 'y'], ax = ax)
    print('count: ', data_df.count())
    all_data_df = pd.concat([all_data_df, data_df])
    
all_data_df.describe()
```
